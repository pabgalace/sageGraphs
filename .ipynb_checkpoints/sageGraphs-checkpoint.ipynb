```python
#Sin peso
Vuelos =Graph({'Madrid':['Barcelona','Palma de Mallorca','Lisboa','Tenerife','Gran Canaria','Londres Heathrow'],'Lisboa':['Madrid'],'Barcelona':['Madrid','Palma de Mallorca','Londres Gatwick','Ámsterdam'],'Palma de Mallorca':['Barcelona','Madrid','Frankfurt'],'Tenerife':['Madrid'],'Gran Canaria':['Madrid'],'Roma':['Palermo','Catania'],'Palermo':['Roma'],'Catania':['Roma'],'Salónca':['Atenas'],'Atenas':['Salónca'],'Toulouse':['París'],'París':['Niza'],'Niza':['París'],'Frankfurt':['Palma de Mallorca','Hamburgo','Berlín'],'Munich':['Düsseldorf','Hamburgo','Berlín'],'Hamburgo':['Frankfurt','Munich'],'Düsseldorf':['Munich','Londres Heathrow'],'Berlín':['Frankfurt','Munich'],'Ámsterdam':['Barcelona','Londres Heathrow'],'Londres Heathrow':['Dublín','Madrid','Düsseldorf','Ámsterdam'],'Londres Gatwick':['Barcelona','Dublín'],'Dublín':['Londres Gatwick','Londres Heathrow'],'Copenhague':['Estocolmo','Oslo'],'Estocolmo':['Copenhague','Oslo'],'Stavenger':['Oslo'],'Bergen':['Oslo'],'Oslo':['Copenhague','Estocolmo','Stavenger','Bergen','Trondheim'],'Trondheim':['Oslo']})
```

```python
Vuelos.show(figsize = 10,aspect_ratio='automatic',vertex_size = 800)
```

```python
print(Vuelos.edges(),"\n")
print("Número de aristas",len(Vuelos.edges()))
```

```python
if Vuelos.is_connected():
    print("El grafo Vuelos es conexo")
else:
        print("Vuelos no es conexo. Tiene ",Vuelos.connected_components_number()," componentes conexas cuyos vértices son:\n")
        for i in [0..Vuelos.connected_components_number()-1]:
            print("Componente número ",i+1,"= ",Vuelos.connected_components()[i],"\n")
```
